```python
def get_header_body(edgar_filepath):
    
    #Open the file
    with open(edgar_filepath, "r") as f:
        preprocess = list(f)
        
    #Get rid of the excess at the top and bottom
    edgar_string = "".join(preprocess[10:-1])
    
    #Get the header
    header_list = re.split("<\/?SEC-HEADER>",edgar_string)
    header = header_list[1]
    
    #Get the body
    edgar_list = re.split("<\/?TEXT>",edgar_string)
    body = ""
    for i in range(len(edgar_list)):
        if i%2==1:
            body = body + edgar_list[i]
            
    #Return the header and body
    return header, body
```

```python
bad_doc_path = "C:\\Users\\zacharybeaver\\Documents\\Github_stuffs\\web_scraping\\edgar\\Parsing_docs\\bad_adds.txt"

def add_document(comp, form_type, cik_num, form_date, doc_header, doc_body):
    
    try:
        #Connect to solr
        s = solr.SolrConnection('http://datamine1:8080/solr')

        #add document to the index
        doc = dict(company = comp, form = form_type, cik = cik_num, date = form_date, header = doc_header, body = doc_body)
        s.add(doc, commit = True)
        return
    
    #if anything goes wrong, write the info out to a file
    except:
        with open(bad_doc_path, "a") as f:
            f.write(comp + "|" + form_type + "|" + cik_num + "|" + form_date + "|" + doc_header + "|" + doc_body + "\n")
        return
```
